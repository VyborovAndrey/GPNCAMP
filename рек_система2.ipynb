```python
import random
# üîπ –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª—É—á–∞–π–Ω—ã—Ö –≤–µ—Å–æ–≤ –¥–ª—è user_answers (—Ç–æ–ª—å–∫–æ –≤–µ—Å–∞, —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –∫—É—Ö–Ω–∏)
user_answers = {
    "wanted_cuisines": {
        "–ï–≤—Ä–æ–ø–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è": round(random.uniform(0, 1), 2), 
        "–ü–∞–Ω–∞–∑–∏–∞—Ç—Å–∫–∞—è –∫—É—Ö–Ω—è": round(random.uniform(0, 1), 2), 
        "–†—É—Å—Å–∫–∞—è –∫—É—Ö–Ω—è": round(random.uniform(0, 1), 2),
        "–ê–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è": round(random.uniform(0, 1), 2), 
        "–ì—Ä—É–∑–∏–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è": round(random.uniform(0, 1), 2)
    },
    "price_limit": {
        500: round(random.uniform(0, 1), 2),
        1000: round(random.uniform(0, 1), 2),
        1500: round(random.uniform(0, 1), 2)
    },
    "food_restrictions": {
        "–ü–æ—Å—Ç–Ω–æ–µ –º–µ–Ω—é": round(random.uniform(0, 1), 2), 
        "–í–µ–≥–µ—Ç–∞—Ä–∏–∞–Ω—Å–∫–æ–µ –º–µ–Ω—é": round(random.uniform(0, 1), 2)
    },
    "walk_time": {
        5: round(random.uniform(0, 1), 2), 
        10: round(random.uniform(0, 1), 2), 
        15: round(random.uniform(0, 1), 2)
    }
}

# üîπ –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∫—É—Ö–Ω–∏ –∏–∑ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞ (–¥–ª—è structured_wishes)
raw_cuisine_list = [
    '–ê–º–µ—Ä–∏–∫–∞–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è; –ï–≤—Ä–æ–ø–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ï–≤—Ä–æ–ø–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è; –§—Ä–∞–Ω—Ü—É–∑—Å–∫–∞—è –∫—É—Ö–Ω—è',
    '–ê–≤—Ç–æ—Ä—Å–∫–∞—è –∫—É—Ö–Ω—è; –ò—Ç–∞–ª—å—è–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ï–≤—Ä–æ–ø–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è; –ü–∞–Ω–∞–∑–∏–∞—Ç—Å–∫–∞—è –∫—É—Ö–Ω—è; –ò—Ç–∞–ª—å—è–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è',
    '–ï–≤—Ä–æ–ø–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è; –†—É—Å—Å–∫–∞—è –∫—É—Ö–Ω—è; –ê–≤—Ç–æ—Ä—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ì—Ä—É–∑–∏–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', 
    '–°—Ä–µ–¥–∏–∑–µ–º–Ω–æ–º–æ—Ä—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ö–∞–≤–∫–∞–∑—Å–∫–∞—è –∫—É—Ö–Ω—è', '–Ø–ø–æ–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–§—Ä–∞–Ω—Ü—É–∑—Å–∫–∞—è –∫—É—Ö–Ω—è',
    '–ö–∏—Ç–∞–π—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ò—Å–ø–∞–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ö–æ—Ä–µ–π—Å–∫–∞—è –∫—É—Ö–Ω—è', '–¢—É—Ä–µ—Ü–∫–∞—è –∫—É—Ö–Ω—è', '–ì—Ä–µ—á–µ—Å–∫–∞—è –∫—É—Ö–Ω—è',
    '–ß–µ—à—Å–∫–∞—è –∫—É—Ö–Ω—è', '–£–∫—Ä–∞–∏–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–í–æ—Å—Ç–æ—á–Ω–∞—è –∫—É—Ö–Ω—è', '–¢–∞–π—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ú–µ–∫—Å–∏–∫–∞–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è',
    '–ê—Ä–º—è–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ò–∑—Ä–∞–∏–ª—å—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ò–Ω–¥–∏–π—Å–∫–∞—è –∫—É—Ö–Ω—è', '–õ–∏–≤–∞–Ω—Å–∫–∞—è –∫—É—Ö–Ω—è', '–ì–∞–≤–∞–π—Å–∫–∞—è –∫—É—Ö–Ω—è'
]

all_cuisines = set()
for cuisines in raw_cuisine_list:
    all_cuisines.update(cuisines.split("; "))

# üîπ –£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –±–ª—é–¥–∞ –∏–∑ –ø—Ä–µ–¥–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω–æ–≥–æ —Å–ø–∏—Å–∫–∞ (–¥–ª—è structured_wishes)
raw_dish_list = [
    '–ü–∞—Å—Ç–∞; –®–≤–µ–¥—Å–∫–∏–π —Å—Ç–æ–ª; –î–µ—Å–µ—Ä—Ç—ã; –ö—É—Ä–∏–Ω—ã–µ –∫—Ä—ã–ª—ã—à–∫–∏; –°–∞–ª–∞—Ç—ã; –î–µ—Ç—Å–∫–æ–µ –º–µ–Ω—é; –ú–∏–¥–∏–∏; –°—É–ø—ã; –ë—É—Ä–≥–µ—Ä—ã; –ü–µ–ª—å–º–µ–Ω–∏; –°—Ç–µ–π–∫-—Ö–∞—É—Å; –°—Ç–µ–π–∫–∏; –ì–æ—Ä—è—á–∏–µ –≤—Ç–æ—Ä—ã–µ –±–ª—é–¥–∞; –ö–∞—Ä—Ç–æ—Ñ–µ–ª—å-—Ñ—Ä–∏',
    '–ü–∞—Å—Ç–∞; –î–µ—Å–µ—Ä—Ç—ã; –ö–æ–Ω–¥–∏—Ç–µ—Ä—Å–∫–∏–µ –∏–∑–¥–µ–ª–∏—è; –ì–ª–∏–Ω—Ç–≤–µ–π–Ω; –ü–æ—Å—Ç–Ω–æ–µ –º–µ–Ω—é; –î–µ—Ç—Å–∫–æ–µ –º–µ–Ω—é; –°—É–ø—ã; –°–∞–ª–∞—Ç—ã; –ì–æ—Ä—è—á–∏–µ –≤—Ç–æ—Ä—ã–µ –±–ª—é–¥–∞',
    '–°—É–ø—ã; –†–∏–º—Å–∫–∞—è –ø–∏—Ü—Ü–∞; –ü–∞—Å—Ç–∞; –ü–∏—Ü—Ü–∞',
    '–î–µ—Å–µ—Ä—Ç—ã; –®–∞—à–ª—ã–∫; –õ—é–ª—è-–∫–µ–±–∞–±; –°—É–ø—ã; –í–∏–Ω–Ω–∞—è –∫–∞—Ä—Ç–∞; –•–∏–Ω–∫–∞–ª–∏; –ß–µ–±—É—Ä–µ–∫–∏; –•–∞—á–∞–ø—É—Ä–∏; –°–∞–ª–∞—Ç—ã; –ì–æ—Ä—è—á–∏–µ –≤—Ç–æ—Ä—ã–µ –±–ª—é–¥–∞; –ö–∞—Ä—Ç–æ—Ñ–µ–ª—å-—Ñ—Ä–∏',
    '–ü–∞—Å—Ç–∞; –ü–æ–∫–µ; –¢–æ–º-–Ø–º; –î–µ—Ç—Å–∫–∞—è –∫–æ–º–Ω–∞—Ç–∞; –°—É—à–∏; –°–∞–ª–∞—Ç—ã; –î–µ—Ç—Å–∫–æ–µ –º–µ–Ω—é; –ü–∏—Ü—Ü–∞; –ë—É—Ä–≥–µ—Ä—ã; –ü–µ–ª—å–º–µ–Ω–∏; –í–∏–Ω–Ω–∞—è –∫–∞—Ä—Ç–∞; –°—Ç–µ–π–∫–∏; –õ–∞–ø—à–∞ wok; –ì–æ—Ä—è—á–∏–µ –≤—Ç–æ—Ä—ã–µ –±–ª—é–¥–∞',
    '–ü–∞—Å—Ç–∞; –î–µ—Å–µ—Ä—Ç—ã; –ì–ª–∏–Ω—Ç–≤–µ–π–Ω; –®–∞—à–ª—ã–∫; –°—É—à–∏; –î–µ—Ç—Å–∫–æ–µ –º–µ–Ω—é; –ü–æ—Å—Ç–Ω–æ–µ –º–µ–Ω—é; –®–∞—É—Ä–º–∞; –ë—É—Ä–≥–µ—Ä—ã; –ü–µ–ª—å–º–µ–Ω–∏; –í–∏–Ω–Ω–∞—è –∫–∞—Ä—Ç–∞; –ß–µ–±—É—Ä–µ–∫–∏; –°–∞–ª–∞—Ç—ã; –ì–æ—Ä—è—á–∏–µ –≤—Ç–æ—Ä—ã–µ –±–ª—é–¥–∞; –ö–æ—Ä–ø–æ—Ä–∞—Ç–∏–≤–Ω—ã–µ –æ–±–µ–¥—ã; –ö–∞—Ä—Ç–æ—Ñ–µ–ª—å-—Ñ—Ä–∏; –í–∞—Ä–µ–Ω–∏–∫–∏'
]

all_dishes = set()
for dishes in raw_dish_list:
    all_dishes.update(dishes.split("; "))

# üîπ –ì–µ–Ω–µ—Ä–∞—Ü–∏—è —Å–ª—É—á–∞–π–Ω—ã—Ö structured_wishes (–¥–æ–ø. –ø–æ–∂–µ–ª–∞–Ω–∏—è)
structured_wishes = {
    "positive_cuisines": random.sample(list(all_cuisines), k=random.randint(1, 3)),
    "negative_cuisines": random.sample(list(all_cuisines), k=random.randint(0, 2)),
    "positive_dishes": random.sample(list(all_dishes), k=random.randint(2, 4))
}

user_answers, structured_wishes
```

```python
import pandas as pd

# üîπ –§—É–Ω–∫—Ü–∏—è –¥–ª—è —É—á–µ—Ç–∞ –ø–æ–∂–µ–ª–∞–Ω–∏–π –ø–æ –±–ª—é–¥–∞–º (–±–µ–∑ —à—Ç—Ä–∞—Ñ–æ–≤ –∑–∞ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ –±–ª—é–¥–∞)
def adjust_score_by_dishes(row, structured_wishes):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –µ—Å—Ç—å –ª–∏ –ø–æ–∑–∏—Ç–∏–≤–Ω—ã–µ –±–ª—é–¥–∞ –≤ –∑–∞–≤–µ–¥–µ–Ω–∏–∏ –∏ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä—É–µ—Ç –±–∞–ª–ª—ã.
    """
    menu = row["Dishes"] if "Dishes" in row and pd.notna(row["Dishes"]) else ""
    menu_items = [dish.strip().lower() for dish in menu.split(";")]  # –†–∞–∑–±–∏–≤–∞–µ–º –Ω–∞ —Å–ø–∏—Å–æ–∫ –±–ª—é–¥

    positive_dish_bonus = 0

    # –ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –∂–µ–ª–∞–µ–º—ã–µ –±–ª—é–¥–∞ –≤ –º–µ–Ω—é
    for dish in structured_wishes["positive_dishes"]:
        if dish.lower() in menu_items:
            positive_dish_bonus += 0.1  # +0.5 –±–∞–ª–ª–∞ –∑–∞ –∫–∞–∂–¥–æ–µ –ª—é–±–∏–º–æ–µ –±–ª—é–¥–æ

    return positive_dish_bonus

# üîπ –§—É–Ω–∫—Ü–∏—è –¥–ª—è —É—á–µ—Ç–∞ –ø–æ–∑–∏—Ç–∏–≤–Ω—ã—Ö –∏ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã—Ö –∫—É—Ö–æ–Ω—å (–∏–∑ `Cuisine`) - –¥–æ–ø. –±–∞–ª–ª—ã –ø–æ structured_wishes
def adjust_cuisine_score(row, structured_wishes):
    """
    –ü—Ä–æ–≤–µ—Ä—è–µ—Ç, –µ—Å—Ç—å –ª–∏ –ø–æ–∑–∏—Ç–∏–≤–Ω—ã–µ –∏–ª–∏ –Ω–µ–≥–∞—Ç–∏–≤–Ω—ã–µ –∫—É—Ö–Ω–∏ –≤ –∑–∞–≤–µ–¥–µ–Ω–∏–∏ –∏ –∫–æ—Ä—Ä–µ–∫—Ç–∏—Ä—É–µ—Ç –±–∞–ª–ª—ã.
    """
    cuisine_list = row["Cuisine"].split(";") if pd.notna(row["Cuisine"]) else []  # –†–∞–∑–±–∏–≤–∞–µ–º —Å—Ç—Ä–æ–∫—É –Ω–∞ —Å–ø–∏—Å–æ–∫

    positive_cuisine_bonus = 0
    negative_cuisine_penalty = 0

    for cuisine in structured_wishes["positive_cuisines"]:
        if cuisine in cuisine_list:
            positive_cuisine_bonus += 0.2  # –£—Å–∏–ª–∏–≤–∞–µ–º –±–∞–ª–ª—ã –∑–∞ –ª—é–±–∏–º—ã–µ –∫—É—Ö–Ω–∏

    for cuisine in structured_wishes["negative_cuisines"]:
        if cuisine in cuisine_list:
            negative_cuisine_penalty -= 0.4  # –®—Ç—Ä–∞—Ñ –∑–∞ –Ω–µ–∂–µ–ª–∞—Ç–µ–ª—å–Ω—ã–µ –∫—É—Ö–Ω–∏

    return positive_cuisine_bonus + negative_cuisine_penalty

# üîπ –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –∏—Ç–æ–≥–æ–≤–æ–≥–æ –±–∞–ª–ª–∞ –∑–∞–≤–µ–¥–µ–Ω–∏—è (–ø–æ –æ–ø—Ä–æ—Å—É + –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ –ø–æ structured_wishes)
def calculate_score(row, user_answers, structured_wishes):
    # üîπ 1. –ë–∞–ª–ª—ã –∑–∞ –∫—É—Ö–Ω—é (–ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞)
    cuisine_score = sum(
        user_answers["wanted_cuisines"].get(c, 0)*2 for c in user_answers["wanted_cuisines"] if row.get(c, False)
    )

    # üîπ 2. –ë–∞–ª–ª—ã –∑–∞ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –ø–æ –ø–∏—Ç–∞–Ω–∏—é (–ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞)
    restrictions_score = sum(
        user_answers["food_restrictions"].get(r, 0) * 1.5 for r in user_answers["food_restrictions"] if row.get(r, False)
    )

    # üîπ 3. –ë–∞–ª–ª—ã –∑–∞ —Ü–µ–Ω—É (–ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞)
    price = row["Average bill"]
    price_score = sum(
        user_answers["price_limit"].get(limit, 0) for limit in user_answers["price_limit"] if pd.notna(price) and price <= limit
    )

    # üîπ 4. –ë–∞–ª–ª—ã –∑–∞ –≤—Ä–µ–º—è –≤ –ø—É—Ç–∏ (–ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞)
    walk_time = min(row["office_1_time"], row["office_2_time"], row["office_3_time"])
    walk_score = sum(
        user_answers["walk_time"].get(limit, 0) for limit in user_answers["walk_time"] if pd.notna(walk_time) and walk_time <= limit
    )

    # üîπ 5. –ë–∞–ª–ª—ã –∑–∞ —Ä–µ–π—Ç–∏–Ω–≥ –∏ –æ—Ç–∑—ã–≤—ã (–ø–æ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º –æ–ø—Ä–æ—Å–∞)
    rating = row["reviews_general_rating"]
    reviews = row["reviews_general_review_count"]
    rating_score = (1 if rating > 4.5 else 0) + (1 if reviews > 200 else 0)

    # üîπ 6. –î–û–ü–û–õ–ù–ò–¢–ï–õ–¨–ù–´–ï –ë–∞–ª–ª—ã –∑–∞ structured_wishes (–∫—É—Ö–Ω–∏ –∏ –±–ª—é–¥–∞)
    structured_cuisine_score = adjust_cuisine_score(row, structured_wishes)
    structured_dish_score = adjust_score_by_dishes(row, structured_wishes)

    # üîπ –û–±—â–∞—è —Å—É–º–º–∞ –±–∞–ª–ª–æ–≤
    total_score = (cuisine_score + restrictions_score + price_score +
                   walk_score + rating_score + structured_cuisine_score + structured_dish_score)

    return total_score



# üîπ –ü—Ä–∏–º–µ–Ω—è–µ–º —Ä–∞—Å—á–µ—Ç –±–∞–ª–ª–æ–≤ –∫–æ –≤—Å–µ–º –∑–∞–≤–µ–¥–µ–Ω–∏—è–º
df["total_score"] = df.apply(lambda row: calculate_score(row, user_answers, structured_wishes), axis=1)

# üîπ –°–æ—Ä—Ç–∏—Ä—É–µ–º –ø–æ –±–∞–ª–ª–∞–º, –∑–∞—Ç–µ–º –ø–æ —Ä–µ–π—Ç–∏–Ω–≥—É, –∑–∞—Ç–µ–º –ø–æ –∫–æ–ª–∏—á–µ—Å—Ç–≤—É –æ—Ç–∑—ã–≤–æ–≤
df_sorted = df.sort_values(by=["total_score", "reviews_general_rating", "reviews_general_review_count"], 
                           ascending=[False, False, False])

# üîπ –í—ã–±–∏—Ä–∞–µ–º —Ç–æ–ø-3 —Ä–µ—Å—Ç–æ—Ä–∞–Ω–∞
top_3_places = df_sorted.head(3)

top_3_places
```

# üîπ –§—É–Ω–∫—Ü–∏—è –¥–ª—è —Ä–∞—Å—á–µ—Ç–∞ –±–∞–ª–ª–æ–≤ –∑–∞–≤–µ–¥–µ–Ω–∏–π
def calculate_score(row):
    # 1Ô∏è‚É£ –ë–∞–ª–ª—ã –∑–∞ –∫—É—Ö–Ω—é (one-hot encoding)
    #–ë–µ—Ä–µ–º –≤—Å–µ –∫—É—Ö–Ω–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–±—Ä–∞–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (wanted_cuisines).
#–ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ —ç—Ç–∞ –∫—É—Ö–Ω—è –≤ –∑–∞–≤–µ–¥–µ–Ω–∏–∏ (row.get(c, False)).

#–°—É–º–º–∏—Ä—É–µ–º –±–∞–ª–ª—ã –∑–∞ –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –∫—É—Ö–Ω–∏
    cuisine_score = sum(
        user_answers["wanted_cuisines"].get(c, 0) for c in user_answers["wanted_cuisines"] if row.get(c, False)
    )

    # 2Ô∏è‚É£ –ë–∞–ª–ª—ã –∑–∞ –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–∏—è –ø–æ –ø–∏—Ç–∞–Ω–∏—é (one-hot encoding)
    #–ë–µ—Ä–µ–º –≤—Å–µ –∫—É—Ö–Ω–∏, –∫–æ—Ç–æ—Ä—ã–µ –≤—ã–±—Ä–∞–ª–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ (wanted_cuisines).
    #–ü—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ —ç—Ç–∞ –∫—É—Ö–Ω—è –≤ –∑–∞–≤–µ–¥–µ–Ω–∏–∏ (row.get(c, False)).
    #–ï—Å–ª–∏ –µ—Å—Ç—å, –¥–æ–±–∞–≤–ª—è–µ–º –µ–µ –≤–µ—Å –≤ –±–∞–ª–ª—ã (get(c, 0)).
     #–°—É–º–º–∏—Ä—É–µ–º –±–∞–ª–ª—ã –∑–∞ –≤—Å–µ –Ω–∞–π–¥–µ–Ω–Ω—ã–µ –∫—É—Ö–Ω–∏
    restrictions_score = sum(
        user_answers["food_restrictions"].get(r, 0) * 2 for r in user_answers["food_restrictions"] if row.get(r, False)

    # 3Ô∏è‚É£ –ë–∞–ª–ª—ã –∑–∞ —Ü–µ–Ω—É 
    price = row["Average bill"]
    if pd.notna(price):  # –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ —Ü–µ–Ω–∞ –Ω–µ NaN
        price_score = sum(
            user_answers["price_limit"].get(limit, 0) for limit in user_answers["price_limit"] if price <= limit
        )
    else:
        price_score = 0  # –ï—Å–ª–∏ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö –æ —Ü–µ–Ω–µ, –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º

    # 4Ô∏è‚É£ –ë–∞–ª–ª—ã –∑–∞ –≤—Ä–µ–º—è –≤ –ø—É—Ç–∏ (–±–µ—Ä–µ–º –º–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∏–∑ —Ç—Ä–µ—Ö –æ—Ñ–∏—Å–æ–≤)
    walk_time = min(row["office_1_time"], row["office_2_time"], row["office_3_time"])
    if pd.notna(walk_time):  # –ü—Ä–æ–≤–µ—Ä–∫–∞, —á—Ç–æ –≤—Ä–µ–º—è –Ω–µ NaN
        walk_score = sum(
            user_answers["walk_time"].get(limit, 0) for limit in user_answers["walk_time"] if walk_time <= limit
        )
    else:
        walk_score = 0  # –ï—Å–ª–∏ –Ω–µ—Ç –¥–∞–Ω–Ω—ã—Ö, –Ω–µ —É—á–∏—Ç—ã–≤–∞–µ–º

    # 5Ô∏è‚É£ –ë–∞–ª–ª—ã –∑–∞ —Ä–µ–π—Ç–∏–Ω–≥ –∏ –æ—Ç–∑—ã–≤—ã
    rating = row["reviews_general_rating"]
    reviews = row["reviews_general_review_count"]
    rating_score = (1 if rating > 4.5 else 0) + (1 if reviews > 200 else 0)

    # üîπ –û–±—â–∞—è —Å—É–º–º–∞ –±–∞–ª–ª–æ–≤
    total_score = cuisine_score + restrictions_score + price_score + walk_score + rating_score

    return total_score
